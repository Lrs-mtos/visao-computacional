```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Carregar a imagem
image = cv2.imread('./imagens/kadabra.jpg')

# Confere se a imagem foi carregada corretamente
if image is None:
    print('Erro ao carregar a imagem')
    exit()

# Mostra a imagem original
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Imagem Original')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python

# Converter para tons de cinza utilizando a fórmula fornecida
gray_image = np.array([[0.3, 0.59, 0.11],[0.3, 0.59, 0.11],[0.3, 0.59, 0.11]])

# Converter para uint8 para visualização
result = cv2.transform(image, gray_image)

# Exibir a imagem em tons de cinza
plt.subplot(121),plt.imshow(result,cmap = 'gray'),plt.title('Result')
plt.xticks([]), plt.yticks([])
plt.show()
```
