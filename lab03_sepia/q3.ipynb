```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

""" R = 0.393 * R + 0.769 * G + 0.189 * B
    G = 0.349 * R + 0.686 * G + 0.168 * B
    B = 0.272 * R + 0.534 * G + 0.131 * B
"""
```

```python
# Carregar a imagem
image = cv2.imread('./imagens/kadabra.jpg')

# Confere se a imagem foi carregada corretamente
if image is None:
    print('Erro ao carregar a imagem')
    exit()

# Mostra a imagem original
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Imagem Original')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def apply_sepia(image):
    # Criar o filtro spia
    sepia_filter = np.array([[0.393, 0.769, 0.189],
                             [0.349, 0.686, 0.168],
                             [0.272, 0.534, 0.131]])
    
    # Aplicar o filtro spia
    sepia_image = cv2.transform(image, sepia_filter)

    return sepia_image.astype(np.uint8)
```

```python
# Aplicar o filtro spia
sepia_image = apply_sepia(image)

# Exibir a imagem spia:
plt.subplot(111),plt.imshow(sepia_image,cmap = 'gray'),plt.title('Imagem Spia')
plt.xticks([]), plt.yticks([])
plt.show()

```
